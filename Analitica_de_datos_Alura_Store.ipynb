## Carga de datos en los DataFrame

```python
ruta_base = "challenge1-data-science-latam/base-de-datos-challenge1-latam/"

data = {
    "tienda_1": pd.read_csv(ruta_base + "tienda_1 .csv"),
    "tienda_2": pd.read_csv(ruta_base + "tienda_2.csv"),
    "tienda_3": pd.read_csv(ruta_base + "tienda_3.csv"),
    "tienda_4": pd.read_csv(ruta_base + "tienda_4.csv")
}

for tienda, df in data.items():
    print(f"{tienda}: {df.shape}")
    display(df.head())
```

## Limpieza y estandarización de datos

```python
dfs = {k: v.copy() for k, v in data.items()}

for name, df in dfs.items():
    df.columns = df.columns.str.lower().str.replace(" ", "_")

    numeric_cols = ["price", "quantity", "review_score", "shipping_time"]
    for col in numeric_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")


    if "price" in df.columns:
        if "quantity" in df.columns:
            df["revenue"] = df["price"] * df.get("quantity", 1).fillna(1)
        else:
            df["revenue"] = df["price"]
    else:
        df["revenue"] = 0  # fallback si no existe price

    print(f" {name} estandarizada")
    display(df.head())
```

## Métricas por tienda

```python
resumen = []

for name, df in dfs.items():

    ingresos = df["precio"].sum()

    reseña = df["calificación"].mean()

    envio = df["costo_de_envío"].mean()

    categoria_top = df["categoría_del_producto"].value_counts().idxmax()

    resumen.append({
        "tienda": name,
        "ingresos_totales": round(ingresos, 2),
        "reseña_promedio": round(reseña, 2),
        "envio_promedio": round(envio, 2),
        "categoria_top": categoria_top
    })

resumen_df = pd.DataFrame(resumen)
display(resumen_df)
```

## Visualizaciones

###Ingresos por tienda

```python
plt.figure(figsize=(7,4))
sns.barplot(data=resumen_df, x="tienda", y="ingresos_totales", color='green')
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingresos Totales (USD)")
plt.show()
```

### Reseñas vs. Ventas

```python
plt.figure(figsize=(7,4))
sns.scatterplot(data=resumen_df, x="reseña_promedio", y="ingresos_totales", hue="tienda", s=150)
plt.title("Relación entre Reseñas y Ventas")
plt.xlabel("Calificación promedio")
plt.ylabel("Ingresos Totales (USD)")
plt.show()
```

### Tiempo de envío promedio

```python
plt.figure(figsize=(7,4))
sns.barplot(data=resumen_df, x="tienda", y="envio_promedio", color ='green')
plt.title("Tiempo de Envío Promedio por Tienda")
plt.ylabel("Tiempo en minutos")
plt.show()
```

### Scatterplot geográfico

```python
plt.figure(figsize=(8,6))

for name, df in dfs.items():
    plt.scatter(df["lon"], df["lat"], s=10, alpha=0.4, label=name)

plt.title("Distribución geográfica de ventas por tienda")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.legend()
plt.show()
```

## Recomendación basada en datos

```python
fig, axes = plt.subplots(2, 2, figsize=(12,10))

axes = axes.flatten()

for ax, (name, df) in zip(axes, dfs.items()):
    hb = ax.hexbin(df["lon"], df["lat"], gridsize=30, cmap="Greens")
    ax.set_title(f"Mapa de calor - {name}")
    ax.set_xlabel("Longitud")
    ax.set_ylabel("Latitud")
    fig.colorbar(hb, ax=ax)

plt.tight_layout()
plt.show()
```
